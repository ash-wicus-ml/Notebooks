```python

# Featurewiz is a brand new Python library that can automatically help you select the best features from your dataset, 
# however big they may be with just a single line of code. There are two major benefits to feature selection:

# You will be able to explain your reasoning better since there are fewer features in your model

# In large data sets, it may be very expensive to build a model with all features since it may take too long to 
# train. However, with featurewiz, you can do it in seconds and with far less features.
```

```python

# import packages 
# let's leverage MTCars...it's a little prosaic, but it will illustrate the point...
import pandas as pd
from featurewiz import featurewiz

from sklearn import datasets
import statsmodels.api as sm
import numpy as np
import pandas as pd

mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df.head()


X = df.drop(['mpg'],axis=1)
y = df.mpg.values 


features = featurewiz(df, target='mpg', corr_limit=0.70, verbose=2)
```

```python

# In the above output, we can clearly see how featurewiz clearly maps different variables with MIS scores 
# and correlation with different feature variables. It is blazingly fast and easy to use. For our dataset, 
# it only took 1 second to generate the output.
```
