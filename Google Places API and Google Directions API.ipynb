```python

import requests

# Set your Google Places API key
api_key = "your_api_key_here"  # Replace with your actual API key

# Google Places API endpoint
places_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Parameters for the API request
params = {
    'location': '40.7454833,-73.9568128',  # Latitude, Longitude (Long Island City, NY)
    'radius': 1000,  # Search within a 1000-meter radius
    'keyword': 'restaurant',  # Search for restaurants
    'key': api_key,
}

# Make the API request
response = requests.get(places_url, params=params)
results = response.json()

# Process and print the results
if 'results' in results:
    for place in results['results']:
        name = place.get('name', 'N/A')
        address = place.get('vicinity', 'N/A')
        print(f"Name: {name}, Address: {address}")
else:
    print("No results found.")
    
```

```python

import requests

# Set your Google Directions API key
api_key = "your_api_key_here"  # Replace with your actual API key

# Google Directions API endpoint
directions_url = "https://maps.googleapis.com/maps/api/directions/json"

# Origin and destination locations
origin = "47-05 Center Blvd, Long Island City, NY"
destination = "60 Wall Street, NYC"

# Parameters for the API request
params = {
    'origin': origin,
    'destination': destination,
    'key': api_key,
}

# Make the API request
response = requests.get(directions_url, params=params)
results = response.json()

# Process and print the results
if 'routes' in results and len(results['routes']) > 0:
    route = results['routes'][0]
    duration = route['legs'][0]['duration']['text']
    distance = route['legs'][0]['distance']['text']

    print(f"Route from {origin} to {destination}:")
    print(f"Duration: {duration}, Distance: {distance}")

    for step in route['legs'][0]['steps']:
        print(step['html_instructions'])
        print("-----")
else:
    print("No results found.")
    
```
