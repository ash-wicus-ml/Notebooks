```python

import csv
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster

df = pd.read_csv ('C:\\Users\\ryans\\Desktop\\Briefcase\\PDFs\\1-ALL PYTHON & R CODE SAMPLES\\Clustering\\Longitude Latitude Coordinates\\cities.csv')
print(df)

cities = df
```

```python

# K-means clustering based on longitude-latitude
kmeans = cluster.KMeans(10)
kmeans.fit(cities[['longitude', 'latitude']])
labels = kmeans.labels_
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.show()
```

```python

# Try different values for K
# Now with centroids
kmeans = cluster.KMeans(5)
kmeans.fit(cities[['longitude', 'latitude']])
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', c='red')
plt.show()
```

```python

# K-means clustering based on temperature
kmeans = cluster.KMeans(5)
kmeans.fit(cities[['temperature']])
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.show()
```
