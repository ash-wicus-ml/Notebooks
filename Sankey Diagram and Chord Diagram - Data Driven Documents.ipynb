```python

# The advantage of d3 is its high performance, deeply customizable charts, and it works with web standards. 
# The name “D3” is short for Data-Driven Documents, which is a Javascript library for producing dynamic, 
# interactive data visualizations, it is fast, beautiful to see, and you can add all your creativity. 
# Furthermore, you don’t need any other technology than a browser to plot the graphs.
```

```python

from d3blocks import D3Blocks

# Initialize
d3 = D3Blocks()

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from pandas import DataFrame


# Intitialise data of lists
data = [{'source': 'Arrivals', 'target':'Greece', 'weight':1590}, 
       {'source': 'Arrivals', 'target':'Spain', 'weight':1455},
       {'source': 'Arrivals', 'target':'France', 'weight':1345}, 
       {'source': 'Arrivals', 'target':'Iceland', 'weight':1100},
       {'source': 'Arrivals', 'target':'Iceland', 'weight':1850},
       {'source': 'Departures', 'target':'Greece', 'weight':2100}, 
       {'source': 'Departures', 'target':'Ireland', 'weight':1000},
       {'source': 'Departures', 'target':'America', 'weight':950}, 
       {'source': 'Departures', 'target':'Ireland', 'weight':1200},
       {'source': 'Departures', 'target':'Japan', 'weight':1050},]
df = pd.DataFrame(data)

df
```

```python

d3.d3graph(df, showfig=False)

# Color node on clustering
d3.D3graph.set_node_properties(color='cluster')


# Show the network graph
d3.D3graph.show()
```

```python

# Everything loooks great so far, but let's try one more scenario, this time, focused on financial-related data. 
# This Sankey diagram contains the complete information of the income statement: The amounts of all items are 
# represented by the width of the streams and the inherent structure of the income statement -- which items are 
# added and how are they grouped -- is represented by how flows divide and join each other.

# That's great! A Sankey diagram is an ideal visualization of an income statement because it can represent the 
# statement's complete information in the simplest possible form.

from d3blocks import D3Blocks

# Initialize
d3 = D3Blocks()

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from pandas import DataFrame

df = pd.read_csv("apple.csv")
df.head()
```

```python

from d3blocks import D3Blocks

# Initialize
d3 = D3Blocks()


# Link settings
d3.sankey(df, link={"color": "source-target"})
```
