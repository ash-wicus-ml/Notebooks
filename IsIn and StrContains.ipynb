```python

import pandas as pd
import numpy as np
data = pd.read_csv('C:\\Users\\titanic.csv')
data.head()
```

```python

# find full string -- finds 'first' or 'third'
df_small = data[data['class'].isin(['First','Third'])]
print(df_small.head())

print('-------------------------------------------------------------------------------')

# find sub-string -- finds NOT 'first' or 'second'
df_small = data.loc[~data['class'].str.contains("fi|se", case=False)]
print(df_small.head())

```

```python

# find sub-string -- finds 'first'
df_small = data.loc[data['class'].str.contains("fi", case=False)]
print(df_small.head())

print('-------------------------------------------------------------------------------')

# find sub-string -- finds 'first' or 'second'
df_small = data.loc[data['class'].str.contains("fi|se", case=False)]
print(df_small.head())
```

```python

# for numbers, we don't use quote marks
age = [35,38,58]
df_small=data[data.age.isin(age)]
print(df_small.head())

print('-------------------------------------------------------------------------------')

# find numbers greater than
df_small = data[data['age'] > 40] 
print(df_small.head())
```

```python

df_small = data['fare'].where(data['fare'] > 50, 0)
print(df_small.head())
```

```python

# find item in column based on two criteria, both being required
df_small = data.loc[(data['age'] >= 22) & (data['age'] <= 32)]
print(df_small.head())
```

```python

data['cumsum_age'] = data[['age','sex','pclass']].groupby('pclass').cumsum()
data.head(10)
```

```python

data['rank'] = data['age'].rank()
data.head(10)
```
